```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('fivethirtyeight')
from matplotlib import rcParams
rcParams['pdf.fonttype'] = 42
%matplotlib inline
```

```python
data = pd.read_excel('./bed.xlsx',sheet_name='Sheet2')
```

```python
data
```

```python
data['rolling'] = data['Total Sales'].rolling(3).mean()
```

```python
fig = plt.figure(figsize=(15,8),dpi=(200))
ax = fig.add_subplot(111)

ax.plot(data['Fiscal Year'],data['Total Sales'])

t = np.polyfit(data.loc[10:,'Fiscal Year'],data.loc[10:,'Total Sales'],1)
tt = np.poly1d(t)
ax.plot(data.loc[10:,'Fiscal Year'],tt(data.loc[10:,'Fiscal Year']),color='green',linestyle="-",linewidth=1)
ax.set_ylim(0,max(data['Total Sales'])*1.1)
fig.savefig('./bed-fy.pdf')
```

```python
fig = plt.figure(figsize=(15,8),dpi=(200))
ax = fig.add_subplot(111)

ax.plot(data['year_int'],data['Total Sales'])

t = np.polyfit(data.loc[10:,'year_int'],data.loc[10:,'Total Sales'],1)
tt = np.poly1d(t)
ax.plot(data.loc[10:,'year_int'],tt(data.loc[10:,'year_int']),color='green',linestyle="-",linewidth=1)
ax.set_ylim(0,max(data['Total Sales'])*1.1)
# fig.savefig('./bed.pdf')
```

```python
fig = plt.figure(figsize=(15,8),dpi=(200))
ax = fig.add_subplot(111)

ax.plot(data['Fiscal Year'],data['Total Sales per capita'])

t = np.polyfit(data.loc[10:,'Fiscal Year'],data.loc[10:,'Total Sales per capita'],1)
tt = np.poly1d(t)
ax.plot(data.loc[10:,'Fiscal Year'],tt(data.loc[10:,'Fiscal Year']),color='green',linestyle="-",linewidth=1)
ax.set_ylim(0,max(data['Total Sales per capita'])*1.1)
fig.savefig('./bed-per-capita-fy.pdf')
```
